```python
import matplotlib.pyplot as plt
import random

# Generate random error values between -2 and 2
n = 20
errors = []                  
for i in range(n):          
    val = random.randint(-2, 2)  # generate a random number between -2 and 2
    errors.append(val) 


mu = 10
phi1 = 0.5
phi2 = 0.5

#  MA(1)
ma1 = []
for i in range(1, len(errors)):
    j = mu + phi1 * errors[i-1]   # depends on previous error
    ma1.append(j)

# MA(2)
ma2 = []
for i in range(2, len(errors)):
    j = mu + phi1 * errors[i-1] + phi2 * errors[i-2]  # depends on last two errors
    ma2.append(j)

plt.figure(figsize=(10,6))
plt.plot(range(1, len(errors)), ma1, marker='s', label="MA(1)")
plt.plot(range(2, len(errors)), ma2, marker='^', label="MA(2)")

plt.title("Moving Average Filters (MA(1) MA(2))")
plt.xlabel("Time index (t)")
plt.ylabel("Filtered Value")
plt.legend()
plt.grid(True)
plt.show()
```
